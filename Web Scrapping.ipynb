<h2>Web Scrapping</h2>    

Web scrapping is used to download data from websites with multiple webpages (that allows us to download data). Python uses a library called bautifulSoup (pip install beautifulSoup4). 

```python
from bs4 import BeautifulSoup
import requests

# give the url you want to download teh data from
url      = "https://stackoverflow.com/questions/43034960/many-to-one-and-many-to-many-lstm-examples-in-keras"

# check if we can download the data
response = requests.get(url)

# if the following prints <Response [200]>, everything is working fine
print(response)
```

```python
'''
the data is downloaded in the form of text with page elements
print data in the code below to see how the data looks like
'''
data = response.text
```

```python
# create a bueatifulSoup object
# it parses the data, however, printing soup does not look much different from data printed above
soup = BeautifulSoup(data,'html.parser')
```

```python
'''
it allows us to get html tags and get data stored in each tag
for example, html stores links in the format <a href=""></a>, so we can get the tag
it will create a list with all the a tags
check and print the links below to see a list of a tags. Since url is stored in href, we use it to get the urls
it will give you all the links that were posted in the website
'''
links = soup.find_all('a')
for link in links:
    print(link.get('href'))
```

```python
'''
EXAMPLE: downloading all the titles from amazon job search page
We always need to print one page atleast to see the structure of the page 
May be search for some words you see on the webpage and check it in soup object, under which tage it comes
'''
url      = "https://www.amazon.jobs/en"
response = requests.get(url)
data     = response.text
soup     = BeautifulSoup(data)

links = soup.find_all('a')
for link in links:
    print(link.get('href'))
```
